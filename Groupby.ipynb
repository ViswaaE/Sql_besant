```python
%%sql
use besant;
```

```python
%%sql
CREATE TABLE sales_data (
    sale_id INT PRIMARY KEY,
    employee_name VARCHAR(50),
    region VARCHAR(50),
    product VARCHAR(50),
    category VARCHAR(50),
    quantity INT,
    price DECIMAL(10,2),
    total_amount DECIMAL(10,2),
    sale_date DATE,
    customer_rating INT
);
```

```python
%%sql
##Show the total sales amount by region.
SELECT REGION , SUM(TOTAL_AMOUNT)
FROM SALES_DATA
GROUP BY REGION;
```

```python
%%sql
##Find the average customer rating for each category.
SELECT CATEGORY, AVG(CUSTOMER_RATING) AS AVG_RATING 
FROM SALES_DATA
GROUP BY CATEGORY;
```

```python
%%sql
##List employees whose total sales exceed 1000.
SELECT EMPLOYEE_NAME , SUM(TOTAL_AMOUNT) AS TOTAL
FROM SALES_DATA
GROUP BY EMPLOYEE_NAME
HAVING SUM(TOTAL_AMOUNT) > 1000;
```

```python
%%sql
##Count the number of distinct products sold in each category.
SELECT CATEGORY , COUNT(DISTINCT PRODUCT) 
FROM SALES_DATA
GROUP BY CATEGORY;  
```

```python
%%sql
##Display products that were sold in more than two transactions.
SELECT PRODUCT , AVG(QUANTITY) AS AVG_QUANTITY
FROM SALES_DATA
group by PRODUCT
HAVING COUNT(*) > 2;
```

```python
%%sql
##Show total sales by each employee in each region.
SELECT REGION,employee_name , SUM(TOTAL_AMOUNT) AS TOTAL_SALES
FROM SALES_DATA
GROUP BY REGION, employee_name;
```

```python
%%sql
##Find regions where the average customer rating is greater than 4.
SELECT REGION,AVG(CUSTOMER_RATING) AS AVG_RATING
FROM SALES_DATA
group by REGION
HAVING AVG(CUSTOMER_RATING) > 4;
```

```python
%%sql
##Show the total quantity of items sold in each category.
SELECT CATEGORY, SUM(QUANTITY) AS TOTAL_QUANTITY_SOLD
FROM SALES_DATA
GROUP BY CATEGORY;
```

```python
%%sql
##Display products that have total sales greater than 1500.
SELECT PRODUCT, SUM(TOTAL_AMOUNT)  AS PRODUCT_AMOUNT
FROM SALES_DATA
GROUP BY PRODUCT
HAVING SUM(TOTAL_AMOUNT) > 1500;
```

```python
%%sql
##Show the average price of products in each category.
select category, product , avg(total_amount) as avg_price
from sales_data
group by product, category;
```
